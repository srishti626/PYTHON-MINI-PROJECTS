```python
import tkinter as tk
from tkinter import messagebox

def add_contact():
    name = e_name.get()
    phone = e_phone.get()
    email = e_email.get()
    address = e_address.get()

    if name and phone:
        contact = {"Name": name, "Phone": phone, "Email": email, "Address": address}
        contacts.append(contact)
        messagebox.showinfo("Success", "Contact added successfully!")
        clear_entries()
    else:
        messagebox.showwarning("Error", "Name and Phone cannot be empty!")

def view_contacts():
    if not contacts:
        messagebox.showinfo("Empty", "No contacts available.")
    else:
        c_list = "\n".join([f"Name: {contact['Name']}, Phone: {contact['Phone']}, Email: {contact['Email']}, Address: {contact['Address']}" for contact in contacts])
        messagebox.showinfo("Contacts", c_list)

def search_contact():
    s_term = e_search.get().lower()
    res = [contact for contact in contacts if s_term in contact['Name'].lower() or s_term in contact['Phone'].lower()]

    if not res:
        messagebox.showinfo("Not Found", "No contacts found.")
    else:
        c_list = "\n".join([f"Name: {contact['Name']}, Phone: {contact['Phone']}, Email: {contact['Email']}, Address: {contact['Address']}" for contact in res])
        messagebox.showinfo("Search Results", c_list)

def clear_entries():
    e_name.delete(0, tk.END)
    e_phone.delete(0, tk.END)
    e_email.delete(0, tk.END)
    e_address.delete(0, tk.END)


root = tk.Tk()
root.geometry("600x400")
root.title("CONTACT MANAGEMENT SYSTEM")

frame = tk.Frame(root, pady=20)
frame.pack()

global contacts
contacts = []

l_name = tk.Label(frame, text="NAME:")
l_name.grid(row=0, column=0, padx=10, pady=5, sticky=tk.E)

e_name = tk.Entry(frame)
e_name.grid(row=0, column=1, padx=10, pady=5)

l_phone = tk.Label(frame, text="PHONE:")
l_phone.grid(row=1, column=0, padx=10, pady=5, sticky=tk.E)

e_phone = tk.Entry(frame)
e_phone.grid(row=1, column=1, padx=10, pady=5)

l_email = tk.Label(frame, text="EMAIL:")
l_email.grid(row=2, column=0, padx=10, pady=5, sticky=tk.E)

e_email = tk.Entry(frame)
e_email.grid(row=2, column=1, padx=10, pady=5)

l_address = tk.Label(frame, text="ADDRESS:")
l_address.grid(row=3, column=0, padx=10, pady=5, sticky=tk.E)

e_address = tk.Entry(frame)
e_address.grid(row=3, column=1, padx=10, pady=5)

btn_add = tk.Button(frame, text="ADD CONTACT", command=add_contact)
btn_add.grid(row=4, column=0, columnspan=2, pady=10)

btn_view = tk.Button(frame, text="VIEW CONTACT", command=view_contacts)
btn_view.grid(row=5, column=0, columnspan=2, pady=10)

l_search = tk.Label(frame, text="SEARCH:")
l_search.grid(row=6, column=0, padx=10, pady=5, sticky=tk.E)

e_search = tk.Entry(frame)
e_search.grid(row=6, column=1, padx=10, pady=5)

btn_search = tk.Button(frame, text="SEARCH", command=search_contact)
btn_search.grid(row=7, column=0, columnspan=2, pady=10)

root.mainloop()
```
